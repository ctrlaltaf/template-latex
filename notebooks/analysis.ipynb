```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sine curve
x = np.linspace(0, 10, 1000)
y = np.sin(x)

plt.figure(figsize=(6,4))
plt.plot(x, y, label="sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Save figure into figs/ for LaTeX
plt.savefig("../figs/sine_curve.pdf")
plt.show()
```
